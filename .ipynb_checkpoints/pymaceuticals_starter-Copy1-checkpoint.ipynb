### Observations and Insights 

* The effectiveness of Capomulin is comparable to Ramicane with slightly better results. Capomulin is far more effective than Infubinol and Ceftamin in reducing tumor size. Additionally, Capomulin and Ramicane were the most consistent treatments as they had the lowest standard errors. 

* Granted, there is a strong, positive correlation between mouse weight and average tumor size with a correlation value of 0.8419 but it is not enough to determine if drug regimen success is factored by weight. 

* The gender of mice tested was near equal, with about 49% female to 51% male. Based on this, there is no gender bias; however, further study is needed to verify survival rates and drug treatment results by gender to determine if there is an unknown factor.

* There was at least one lucky mouse taking Infubinol where it had a tumor volume reduction well below the statistical lower bound for the drug. This should not be taken under consideration to retest the drug though. 

```python
# dependency setup
import matplotlib.pyplot as plt
import pandas as me
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# read mouse data and  study results
mouseMetaData = me.read_csv(mouse_metadata_path)
studyResults = me.read_csv(study_results_path)

# combine to single dataset
combinedPharmaResults = me.merge(studyResults, mouseMetaData, how = 'left', on=['Mouse ID', 'Mouse ID'])

# display preview
combinedPharmaResults.head()
```

### Summary Statistics

```python
# summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen using
# groupby and summary statistical methods to calculate the following properties of each drug regimen: mean, median, variance,
# standard deviation, and SEM of the tumor volume. 

# group drugs together
drugGroups = resultsDF.groupby(['Drug Regimen'])

# regimen statistical summary
# regimen mean and median
regimenMean = drugGroups['Tumor Volume (mm3)'].mean()
regimenMedian = drugGroups['Tumor Volume (mm3)'].median()

# get variance for entire population
regimenVariance = drugGroups['Tumor Volume (mm3)'].var()

# get standard deviation
regimenStdev = drugGroups['Tumor Volume (mm3)'].std()

# get standard error of means
regimenSEM = drugGroups['Tumor Volume (mm3)'].sem()

# create dataframe to clean up regimen results
regimenStats_a = me.DataFrame({
    'Mean Tumor Volume': regimenMean,
    'Median Tumor Volume': regimenMedian,
    'Tumor Volume Variance': regimenVariance,
    'Tumor Volume Std. Dev.': regimenStdev,
    'Tumor Volume Std. Err.': regimenSEM 
})

# show results
regimenStats_a
```

```python
# for each regimen using the aggregation method, produce the same summary statistics in a single line

# aggregate function for summary stats
regimenStats_b = drugGroups['Tumor Volume (mm3)'].agg(['mean','median', 'var','std', 'sem'])

# create dataframe to clean up regimen results
regimenStats_b = regimenStats_b.rename(columns={'mean' : 'Mean Tumor Volume',
                                                'median' : 'Median Tumor Volume',
                                                'var' : 'Tumor Volume Variance',
                                                'std' : 'Tumor Volume Std. Dev', 
                                                'sem' : 'Tumor Volume Std. Err.'})

# show results
regimenStats_b
```

### Quartiles, Outliers and Boxplots

```python
# calculating the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# get greatest (last) timepoint for each mouse
lastTime = resultsDF.groupby(['Mouse ID']).max()
lastTimeReset = lastTime.reset_index()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
mergedData = lastTimeReset[['Mouse ID', 'Timepoint']].merge(resultsDF, on=['Mouse ID', 'Timepoint'], how='left')
mergedData.head()
```

```python
# treatment list for for loop (and later for plot labels)
drugList = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# list to store tumor values for boxplot
tumorValList = []

# calculate IQR and quantitatively determine if there are any potential outliers
# loop through drug list by pulling in tumor values from merged data
for drug in drugList:
    # get data from dataframe
    tumorVols = mergedData.loc[mergedData['Drug Regimen'] == drug]['Tumor Volume (mm3)']
    
    # append value to tumor value list
    tumorValList.append(tumorVols)
    
    # get statistal data
    quartiles = tumorVols.quantile([.25,.5,.75])
    firstq = round(quartiles[0.25], 4)
    lastq = round(quartiles[0.75], 4)
    iqr = round((lastq - firstq), 4)
    lowerbound = round(firstq - (1.5 * iqr), 4)
    upperbound = round(lastq + (1.5 * iqr), 4)
    
    # determine outliers
    outliers = tumorVols.loc[(tumorVols < lowerbound) | (tumorVols > upperbound)]
    outliersCount = outliers.count()

    # display complete sumamry per drug in list
    print(f"-------------------------------------------")
    print(f"The lower quartile for {drug} is {firstq}.")
    print(f"The upper quartile for {drug} is {lastq}.")
    print(f"The interquartile range for {drug} is {iqr}.")
    print(f"Values below {lowerbound} could be outliers for {drug}.")
    print(f"Values above {upperbound} could be outliers for {drug}.")
    
    # different display for outliers if count > 0
    if outliersCount == 0:
        print(f"{drug} has no potential outliers.") 
    elif outliersCount > 0:
        print(f"{drug}'s potential outliers are: {outliers}.")

    
```

```python
# box plot of the final tumor volume of each mouse across four regimens of interest
fig1, ax = plt.subplots()

# dictionary to set marker
flierprops = dict(marker='o', markerfacecolor='tab:cyan', markersize=8, linestyle='none', markeredgecolor='k')
medianprops = ('tab:blue')

# set data labels
labels = drugList

ax.set_title("Treatment Results", fontweight='bold', fontsize=14)
ax.set_xlabel('Drug Regimen', fontweight='bold', fontsize=12)
ax.set_ylabel('Final Tumor Volume (mm3)', fontweight='bold', fontsize=12)


ax.boxplot(tumorValList, labels=labels, flierprops=flierprops)
plt.show()
```

```python
# scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

# all Capomulin mice are in capMice from previous line plot
capMiceGroup = capMice.groupby(['Mouse ID']).mean()

# get average tumor volume and weights
avgVol = capMiceGroup['Tumor Volume (mm3)']
weights = capMiceGroup['Weight (g)']

plt.scatter(weights, avgVol, marker='o', facecolors='tab:cyan', edgecolors='k')
plt.xlabel('Weight (g)', fontweight='bold', fontsize=12)
plt.ylabel('Avg. Tumor Volume (mm3)', fontweight='bold', fontsize=12)
plt.title('Average Tumor Volume vs. Mouse Weight', fontweight='bold', fontsize=14)
plt.show()
```

### Correlation and Regression

```python
# calculated the correlation coefficient and linear regression model # for mouse weight and average tumor volume for 
# the Capomulin regimen
corr = st.pearsonr(avgVol, weights)
xVals = weights
yVals = avgVol
(slope, intercept, rvalue, pvalue, stderr) = linregress(xVals, yVals)
regressVals = xVals * slope + intercept
lineEQ = 'y = ' + str(round(slope,2)) + 'x + ' + str(round(intercept,2))

print(f"The correlation between mouse weight and average tumor volume is {round(corr[0],4)}.")
print(f"The r-squared value is: {round(rvalue**2, 4)}.")

plt.scatter(xVals, yVals, marker='o', facecolors='tab:cyan', edgecolors='k')
plt.plot(xVals, regressVals, 'c-', alpha=0.5)
plt.annotate(lineEQ, xy=(21,36), fontsize=12, color='tab:blue')
plt.xlabel('Weight (g)', fontweight='bold', fontsize=12)
plt.ylabel('Avg. Tumor Volume (mm3)', fontweight='bold', fontsize=12)
plt.title('Average Tumor Volume vs. Mouse Weight', fontweight='bold', fontsize=14)
plt.show()
```
